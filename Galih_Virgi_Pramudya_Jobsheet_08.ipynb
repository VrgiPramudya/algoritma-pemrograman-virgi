#Praktikum 04 - Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

#Praktikum 07 - Fungsi Decorator

```python
def decorator_sapa(fungsi):
	def wrapper():
		print("Sebelum fungsi dijalankan...")
		fungsi()
		print("Setelah fungsi dijalankan")
	return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines")

sapa()
```

#Penugasan No. 10 - Fungsi Decorator

```python
def logger(fungsi):
    def wrapper():
        print("Memulai fungsi...")
        fungsi()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo, selamat datang!")

halo()
```

#Penugasan No. 11 - Studi Kasus Nyata: Program Nilai

```python
def input_nilai():
    nilai = []
    jumlah = int(input("Masukkan jumlah nilai: "))
    for i in range(jumlah):
        n = float(input(f"Nilai ke-{i+1}: "))
        nilai.append(n)
    return nilai

def hitung_rata(nilai):
    if len(nilai) == 0:
        return 0
    return sum(nilai) / len(nilai)

def status_kelulusan(rata):
    return "Lulus" if rata >= 60 else "Tidak Lulus"

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

nilai_mahasiswa = input_nilai()
rata = hitung_rata(nilai_mahasiswa)

print("\nRata-rata nilai:", rata)
print("Status:", status_kelulusan(rata))

tugas = rata
ujian = float(input("Masukkan nilai ujian: "))

print("Nilai akhir (40% tugas + 60% ujian):", nilai_akhir(tugas, ujian))
```
