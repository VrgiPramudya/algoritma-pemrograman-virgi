# Penugasan No. 6 - Modul Buatan Sendiri

```python
%%writefile konversi.py
def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

```python
%%writefile main_konversi.py

import konversi

c = 30
f = 86

print(f"{c}°C = {konversi.celcius_ke_fahrenheit(c):.2f}°F")
print(f"{f}°F = {konversi.fahrenheit_ke_celcius(f):.2f}°C")
```

# Penugasan No. 8 - Studi Kasus Mini

```python
%%writefile laporan.py
def ringkasan(score, total):
    print("\n===== RINGKASAN NILAI =====")
    print(f"Total Soal : {total}")
    print(f"Jawaban Benar : {score}")
    print(f"Nilai Akhir : {(score/total)*100:.2f}")
    print("============================")
```

```python
import random
import laporan

jumlah_soal = 5
score = 0

print("=== Kuis Penjumlahan Sederhana ===")

for i in range(1, jumlah_soal + 1):
    a = random.randint(1, 20)
    b = random.randint(1, 20)

    print(f"Soal {i}: {a} + {b} = ?")
    jawaban = int(input("Jawab: "))

    if jawaban == a + b:
        print("Benar!\n")
        score += 1
    else:
        print(f"Salah, jawaban yang benar: {a + b}\n")

laporan.ringkasan(score, jumlah_soal)
```
