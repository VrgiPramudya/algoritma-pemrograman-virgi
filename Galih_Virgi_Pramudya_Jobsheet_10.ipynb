#Praktikum 01 - Bubble Sort

```python
def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j] > data[j + 1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

#Praktikum 02 - Selection Sort

```python
def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = i
    for j in range(i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

#Praktikum 03 - Insertion Sort

```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
      key = data[i]
      j = i - 1
      while j >= 0 and data[j] > key:
        data[j + 1] = data[j]
        j -= 1
      data[j + 1] = key
    return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

#Praktikum 04 - Studi Kasus: Mengurutkan Nilai Mahasiswa

```python
def bubble_sort_nilai(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j][1] < data[j + 1][1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

#Praktikum 06 - Sorting Data Barang (Nama & Harga)

```python
def bubble_sort_key(data, key_index=0, reverse=False):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if reverse:
        if data[j][key_index] < data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
      else:
        if data[j][key_index] > data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

#Penugasan No. 4 -	Eksperimen Selection Sort

```python
def selection_sort_desc(data):
    n = len(data)
    for i in range(n - 1, -1, -1):
        max_index = 0
        for j in range(1, i + 1):
            if data[j] > data[max_index]:
                max_index = j

        data[i], data[max_index] = data[max_index], data[i]

    return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort (desc):", selection_sort_desc(angka))
```

#Penugasan No. 5 - Eksperimen Insertion Sort

```python
def insertion_sort_string(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1

        while j >= 0 and data[j].lower() > key.lower():
            data[j + 1] = data[j]
            j -= 1

        data[j + 1] = key

    return data


# Input data string (nama mahasiswa)
jumlah = int(input("Masukkan jumlah mahasiswa: "))
mahasiswa = []

for i in range(jumlah):
    nama = input(f"Nama mahasiswa ke-{i+1}: ")
    mahasiswa.append(nama)

print("\nData sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan alfabetis:", insertion_sort_string(mahasiswa))
```

#Penugasan No. 7 - Analisis Kompleksitas

```python
import random

def bubble_count(data):
    n = len(data)
    count = 0
    for i in range(n):
        for j in range(0, n - i - 1):
            count += 1
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return count

def selection_count(data):
    n = len(data)
    count = 0
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            count += 1
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return count

def insertion_count(data):
    n = len(data)
    count = 0
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0:
            count += 1
            if data[j] > key:
                data[j + 1] = data[j]
                j -= 1
            else:
                break
        data[j + 1] = key
    return count

sizes = [10, 50, 100]

for size in sizes:
    data = [random.randint(1, 1000) for _ in range(size)]

    bubble = bubble_count(data.copy())
    selection = selection_count(data.copy())
    insertion = insertion_count(data.copy())

    print(f"\nUkuran data: {size}")
    print(f"Bubble Sort     : {bubble} perbandingan")
    print(f"Selection Sort  : {selection} perbandingan")
    print(f"Insertion Sort  : {insertion} perbandingan")
```

#Penugasan No. 8 - Proyek Mini

```python
def insertion_sort(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1

        while j >= 0:
            if reverse:
                if data[j][key_index] < key[key_index]:
                    data[j + 1] = data[j]
                else:
                    break
            else:
                if data[j][key_index] > key[key_index]:
                    data[j + 1] = data[j]
                else:
                    break
            j -= 1

        data[j + 1] = key

    return data

#Program utama
barang = []

print("=== INPUT DATA BARANG ===")
jumlah = int(input("Masukkan jumlah barang: "))

for i in range(jumlah):
    print(f"\nBarang ke-{i+1}:")
    nama = input("Nama barang : ")
    harga = int(input("Harga       : "))
    barang.append((nama, harga))

print("\n=== PILIHAN PENGURUTAN ===")
print("1. Urut berdasarkan Nama")
print("2. Urut berdasarkan Harga")

pilih = int(input("Pilih opsi (1/2): "))

print("\nPilih urutan:")
print("1. Ascending (A → Z / murah → mahal)")
print("2. Descending (Z → A / mahal → murah)")

urut = int(input("Pilih (1/2): "))

key_index = 0 if pilih == 1 else 1
reverse = True if urut == 2 else False

hasil = insertion_sort(barang.copy(), key_index, reverse)

print("\n=== HASIL PENGURUTAN ===")
for nama, harga in hasil:
    print(f"{nama} - Rp{harga}")
```
