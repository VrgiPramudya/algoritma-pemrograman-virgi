#Praktikum 01 - Sequential Search pada Data Angka

```python
def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data {target} ditemukan di indeks {i}"
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

#Praktikum 02 - Binary Search pada Data Angka

```python
def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Data {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

#Praktikum 03 - Sequential Search untuk Data Nama

```python
def sequential_search_nama(data, target):
  for i in range(len(data)):
    if data[i].lower() == target.lower():
      return f"Nama {target} ditemukan di indeks {i}"
  return f"Nama {target} tidak ditemukan"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```

#Praktikum 04 - Studi Kasus Nilai Mahasiswa dengan Binary Search

```python
def binary_search_nilai(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Nilai {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Nilai {target} tidak ditemukan"

nilai = [60, 70, 75, 80, 85, 90]
print("Data nilai:", nilai)
print(binary_search_nilai(nilai, 85))
print(binary_search_nilai(nilai, 65))
```

#Praktikum 05 - Perbandingan Waktu Sequential vs Binary Search

```python
import time

def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return i
  return -1

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1

data = list(range(1, 100001))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

#Praktikum 06 - Kombinasi Sorting dan Searching

```python
import random

def insertion_sort(data):
  for i in range(1, len(data)):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"{target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"{target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data setelah diurutkan:", urut)
target = urut[7] # ambil angka ke-8 sebagai target
print(binary_search(urut, target))
```

#Penugasan No. 3 - Variasi Sequential Search

```python
def sequential_search_count(data, target):
  comparisons = 0
  for i in range(len(data)):
    comparisons += 1
    if data[i] == target:
      return f"Data {target} ditemukan di indeks {i} setelah {comparisons} perbandingan"
  return f"Data {target} tidak ditemukan setelah {comparisons} perbandingan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search_count(angka, 30))
print(sequential_search_count(angka, 100))
```

#Penugasan No. 4 - Variasi Binary Search

```python
def binary_search_steps(data, target):
  left, right = 0, len(data)-1
  steps = 0

  while left <= right:
    steps += 1
    mid = (left + right) // 2

    if data[mid] == target:
      return f"Data {target} ditemukan di indeks {mid} setelah {steps} langkah"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1

  return f"Data {target} tidak ditemukan setelah {steps} langkah"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search_steps(angka, 40))
print(binary_search_steps(angka, 100))
```

#Penugasan No. 6 - Studi Kasus Barang Toko

```python
def insertion_sort_by_price(data):
    #Mengurutkan data berdasarkan harga (ascending)
    for i in range(1, len(data)):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j][1] > key[1]:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data


def binary_search_price(data, target):
    left, right = 0, len(data) - 1

    while left <= right:
        mid = (left + right) // 2
        harga_mid = data[mid][1]

        if harga_mid == target:
            return f"Barang dengan harga {target} adalah '{data[mid][0]}' (indeks {mid})"
        elif harga_mid < target:
            left = mid + 1
        else:
            right = mid - 1

    return f"Tidak ada barang dengan harga {target}"


#Daftar barang
barang = [
    ("Pensil", 3000),
    ("Buku", 5000),
    ("Penghapus", 2000),
    ("Pulpen", 4000),
    ("Spidol", 7000)
]

print("Daftar barang sebelum diurutkan:")
print(barang)

#Urutkan berdasarkan harga
urut = insertion_sort_by_price(barang)

print("\nDaftar barang setelah diurutkan berdasarkan harga:")
print(urut)

#Cari harga tertentu
target = int(input("\nMasukkan harga barang yang ingin dicari: "))

print(binary_search_price(urut, target))
```

#Penugasan No. 7 - Analisis Kompleksitas

```python
import time

def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1


def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


def test_search(size):
    data = list(range(size))    #data sudah terurut
    target = size - 1           #elemen terakhir (yang paling lama dicari)

    start = time.time()
    sequential_search(data, target)
    end = time.time()
    sequential_time = end - start

    start = time.time()
    binary_search(data, target)
    end = time.time()
    binary_time = end - start

    return sequential_time, binary_time


sizes = [100, 1000, 10000]

print("Ukuran Data | Sequential Search (detik) | Binary Search (detik)")
print("-------------------------------------------------------------")

for size in sizes:
    seq_time, bin_time = test_search(size)
    print(f"{size:<12} | {seq_time:<24.10f} | {bin_time:<.10f}")
```

#Penugasan No. 8 - Proyek Mini

```python
def insertion_sort(data):
    for i in range(1, len(data)):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j][1] > key[1]:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data


def binary_search(data, target_nilai):
    left, right = 0, len(data) - 1

    while left <= right:
        mid = (left + right) // 2
        if data[mid][1] == target_nilai:
            return mid
        elif data[mid][1] < target_nilai:
            left = mid + 1
        else:
            right = mid - 1
    return -1


mahasiswa = []
jumlah = int(input("Berapa jumlah mahasiswa? "))

for _ in range(jumlah):
    nama = input("Nama: ")
    nilai = int(input("Nilai: "))
    mahasiswa.append((nama, nilai))

print("\nData sebelum diurutkan:")
print(mahasiswa)

mahasiswa = insertion_sort(mahasiswa)

print("\nData setelah diurutkan berdasarkan nilai:")
for m in mahasiswa:
    print(m)

cari = int(input("\nMasukkan nilai yang ingin dicari: "))
index = binary_search(mahasiswa, cari)

if index != -1:
    print(f"Nilai {cari} ditemukan! Pemilik nilai: {mahasiswa[index][0]} (indeks {index})")
else:
    print(f"Nilai {cari} tidak ditemukan dalam daftar.")
```
